# Python Modules

Example from Python documentation

#### ```fibo.py```


```

def fib(n):    # write Fibonacci series up to n
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()

def fib2(n):   # return Fibonacci series up to n
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a+b
    return result
    
if __name__ == "__main__":
    import sys
    fib(int(sys.argv[1]))
```

````python
# Importing fib and fib2 module from fibo.py file
from fibo import fib, fib2

print(fib(500))
print(fib2(500))
````
